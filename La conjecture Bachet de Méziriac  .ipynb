```python
#La conjecture Bachet de Méziriac  

#i) 1ère version
from math import sqrt
compteur = 0
n = 2020
A = int(sqrt(n))
for a in range(A, -1 ,-1):
    reste1 = n - a**2
    B = int(sqrt(reste1))
    
    for b in range(B, -1, -1):
        reste2 = reste1 - b**2
        C = int(sqrt(reste2))
        
        for c in range(C, -1,-1):
            reste3 = reste2 - c**2
            D = int(sqrt(reste3))
            
            for d in range(D, -1, -1):
                if a**2 + b**2 + c**2 + d**2 == n and compteur == 0:
                    print("a =", a, "b =", b, "c =", c , "d =", d)
                    compteur = 1
```

```python
#i) 2ème version
from math import sqrt
compteur = 0
n = 2020
A = int(sqrt(n))
for a in range(A, -1 ,-1):
    reste1 = n - a**2
    B = int(sqrt(reste1))
    
    for b in range(B, -1, -1):
        reste2 = reste1 - b**2
        C = int(sqrt(reste2))
        
        for c in range(C, -1,-1):
            reste3 = reste2 - c**2
            D = int(sqrt(reste3))

            if reste3 == D**2 and compteur == 0: 
                print("a =", a, "b =", b, "c =", c , "d =", D) 
                compteur = 1
```
