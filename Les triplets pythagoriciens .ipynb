```python
#Les triplets pythagoriciens 
#i) 1ère version
for a in range(1, 2020):
    for b in range(1, 2020):
        if a**2 + b**2 == 2020**2 :
            print("a = %d \t b = %d" %(a,b))
```

```python
#ii) 2ème version
for a in range(1,2020):
    ecart = sqrt(2020**2 - a**2)
    if ecart == int(ecart):
        print("a = %d \t b = %d" %(a, int(ecart)))
```
